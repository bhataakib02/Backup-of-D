```python
import sys
sys.path.append('..')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from config import *
```

## Load and Explore Dataset


```python
# Load labeled dataset
df = pd.read_csv(LABELED_DATASET_PATH)
print(f"Total requirements: {len(df)}")
df.head()
```

```python
# Conflict distribution
conflict_counts = df[CONFLICT_TYPES].sum()
print("Conflict Distribution:")
print(conflict_counts)

plt.figure(figsize=(10, 6))
conflict_counts.plot(kind='bar')
plt.title('Conflict Type Distribution')
plt.ylabel('Count')
plt.xlabel('Conflict Type')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
